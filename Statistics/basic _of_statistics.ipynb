# Data and Sampling Distributions

## Sample
A subset from a larger data set.

## Population
- Definition: The population refers to the entire set of individuals, items, or data points of interest that you want to study or make inferences about.

- The larger data set or idea of a data set.

## N (n)
The size of the population (sample).

## Random sampling
Drawing elements into a sample at random.

## Stratified sampling
Dividing the population into strata and randomly sampling from each strata.

## Stratum (pl., strata)
A homogeneous subgroup of a population with common characteristics.

## Simple random sample
The sample that results from random sampling without stratifying the population.

## Bias
Systematic error.

## Sample bias
A sample that misrepresents the population.


# Bias
Statistical bias refers to measurement or sampling errors that are systematic and pro
duced by the measurement or sampling process. An important distinction should be
made between errors due to random chance and errors due to bias
# Size Versus Quality: When Does Size Matter?
In the era of big data, it is sometimes surprising that smaller is better. Time and effort
spent on random sampling not only reduces bias but also allows greater attention to
data exploration and data quality. For example, missing data and outliers may contain
useful information. It might be prohibitively expensive to track down missing values
or evaluate outliers in millions of records, but doing so in a sample of several thou
sand records may be feasible. Data plotting and manual inspection bog down if there
is too much data
* Even in the era of big data, random sampling remains an important arrow in the
data scientists quiver.
* Bias occurs when measurements or observations are systematically in error
because they are not representative of the full population.
* Data quality is often more important than data quantity, and random sampling
can reduce bias and facilitate quality improvement that would otherwise be pro
hibitively expensive.

## Sample statistic
A metric calculated for a sample of data drawn from a larger population.

## Data distribution
The frequency distribution of individual values in a data set.

## Sampling distribution
The frequency distribution of a sample statistic over many samples or resamples.

## Central limit theorem
The tendency of the sampling distribution to take on a normal shape as sample size rises.

## Standard error
The variability (standard deviation) of a sample statistic over many samples (not to be confused with standard deviation, which by itself, refers to variability of individual data values).


# The Python code uses seaborns FacetGrid to show the three histograms:


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Defining a dummy dataset for loans_income
loans_income = pd.Series(np.random.randint(20000, 120000, 10000))

# Sample data
sample_data = pd.DataFrame({
    'income': loans_income.sample(1000),
    'type': 'Data',
})

# Mean of 5 samples
sample_mean_05 = pd.DataFrame({
    'income': [loans_income.sample(5).mean() for _ in range(1000)],
    'type': 'Mean of 5',
})

# Mean of 20 samples
sample_mean_20 = pd.DataFrame({
    'income': [loans_income.sample(20).mean() for _ in range(1000)],
    'type': 'Mean of 20',
})

# Concatenating results
results = pd.concat([sample_data, sample_mean_05, sample_mean_20])

# Creating the FacetGrid
g = sns.FacetGrid(results, col='type', col_wrap=1, height=2, aspect=2)
g.map(plt.hist, 'income', range=[0, 200000], bins=40)

# Setting axis labels and titles
g.set_axis_labels('Income', 'Count')
g.set_titles('{col_name}')

# Display the plot
plt.show()
```

## Central Limit Theorem
* The phenomenon weve just described is termed the central limit theorem. It says that the means drawn from multiple samples will resemble the familiar bell-shaped normal curve, even if the source population is not normally distributed, provided that the sample size is large enough and the departure of the data from normality is not too great. The central limit theorem allows normal-approximation formulas like the t-distribution to be used in calculating sampling distributions for inferencethat is, confidence intervals and hypothesis tests.

* The central limit theorem receives a lot of attention in traditional statistics texts because it underlies the machinery of hypothesis tests and confidence intervals, which themselves consume half the space in such texts.
## central limit theoram assumpation
- 1 The sample must be independent 
- 2 Sample must be indentically distribution
- 3-sample size must be constant

## Standard Error
The standard error is a single metric that sums up the variability in the sampling distribution for a statistic. The standard error can be estimated using a statistic based on the standard deviation \( s \) of the sample values, and the sample size \( n \):

Standard error =  SE = s/sqrt{n}


## Bootstrap sample
A sample taken with replacement from an observed data set.

## Resampling
The process of taking repeated samples from observed data; includes both bootstrap and permutation (shuffling) procedures.


```python
from sklearn.utils import resample

# Defining a dummy dataset for loans_income
loans_income = pd.Series(np.random.randint(20000, 120000, 10000))
loans_income
```

```python
# Performing bootstrap procedure
results = []
for nrepeat in range(1000):
    sample = resample(loans_income)
    results.append(sample.median())
results = pd.Series(results)
```

```python
# Printing bootstrap statistics
print('Bootstrap Statistics:')
print(f'Original Median: {loans_income.median()}')
print(f'Bias: {results.mean() - loans_income.median()}')
print(f'Standard Error: {results.std()}')
```

```python

from sklearn.utils import resample

# Defining a dummy dataset for loans_income
loans_income = pd.Series(np.random.randint(20000, 120000, 10000))

# Performing bootstrap procedure
results = []
for nrepeat in range(1000):
    sample = resample(loans_income)
    results.append(sample.median())
results = pd.Series(results)

# Printing bootstrap statistics
print('Bootstrap Statistics:')
print(f'Original Median: {loans_income.median()}')
print(f'Bias: {results.mean() - loans_income.median()}')
print(f'Standard Error: {results.std()}')
```

## Imports
`resample` from `sklearn.utils` is used to create bootstrap samples.

## Bootstrap Procedure
- **Loop:** Generates 1000 bootstrap samples from `loans_income`, calculates the median of each sample, and stores it in the `results` list.
- **Conversion:** Converts the list of medians to a pandas `Series`.

## Statistics Calculation
- **Original Median:** The median of the original dataset.
- **Bias:** Difference between the mean of the bootstrap medians and the original median.
- **Standard Error:** Standard deviation of the bootstrap medians, providing an estimate of the variability of the median.
- The bootstrap (sampling with replacement from a data set) is a powerful tool for assessing the variability of a sample statistic.
- The bootstrap can be applied in a similar fashion in a wide variety of circumstances, without extensive study of mathematical approximations to sampling distributions.
- It also allows us to estimate sampling distributions for statistics where no mathematical approximation has been developed.
- When applied to predictive models, aggregating multiple bootstrap sample predictions (bagging) outperforms the use of a single model.

Distributions of sample means are typically shaped like a t-distribution, and there is a family of t-distributions that differ depending on how large the sample is. The larger the sample, the more normally shaped the t-distribution becomes.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define degrees of freedom for the t-distribution
df = 10

# Generate a range of x values for plotting
x = np.linspace(-5, 5, 1000)

# Generate the t-distribution PDF
pdf = stats.t.pdf(x, df)

# Plot the t-distribution
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f't-distribution (df={df})')
plt.title('t-Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

## Binomial Distribution
Yes/no (binomial) outcomes lie at the heart of analytics since they are often the culmination of a decision or other process; buy/dont buy, click/dont click, survive/die, and so on. Central to understanding the binomial distribution is the idea of a set of trials, each trial having two possible outcomes with definite probabilities.


## Key Terms for Binomial Distribution

- **Trial**  
  An event with a discrete outcome (e.g., a coin flip).

- **Success**  
  The outcome of interest for a trial.  
  **Synonym:** 1 (as opposed to 0)

- **Binomial**  
  Having two outcomes.  
  **Synonyms:** yes/no, 0/1, binary

- **Binomial trial**  
  A trial with two outcomes.  
  **Synonym:** Bernoulli trial

- **Binomial distribution**  
  Distribution of the number of successes in \( x \) trials.  
  **Synonym:** Bernoulli distribution

## Additional Information

- Binomial outcomes are important to model, since they represent, among other things, fundamental decisions (buy or dont buy, click or dont click, survive or die, etc.).
- A binomial trial is an experiment with two possible outcomes: one with probability \( p \) and the other with probability \( 1 - p \).
- With large \( n \), and provided \( p \) is not too close to 0 or 1, the binomial distribution can be approximated by the normal distribution.


```python
import scipy.stats as stats

# Define parameters for the binomial distribution
n = 20  # Number of trials
p = 0.5  # Probability of success

# Generate a range of x values (number of successes)
x = np.arange(0, n + 1)

# Calculate the binomial distribution PMF
pmf = stats.binom.pmf(x, n, p)

# Plot the binomial distribution
plt.figure(figsize=(8, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.xticks(x)
plt.show()
```

- `p` specifies the probability of success in each trial.

## Calculate PMF
- `stats.binom.pmf(x, n, p)` computes the probability mass function (PMF) for each value in \( x \).

## Plot
- `plt.bar(x, pmf, color='skyblue', edgecolor='black')` creates a bar plot of the PMF.

# Chi-Square Distribution

category counts
-  The chi-square distribution is typically concerned with counts of subjects or

- The chi-square statistic measures the extent of departure from what you would

```python
# Define parameters for the chi-square distribution
df = 10  # Degrees of freedom

# Generate a range of x values for plotting
x = np.linspace(0, 30, 1000)

# Calculate the chi-square distribution PDF
pdf = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'Chi-Square Distribution (df={df})', color='blue')
plt.fill_between(x, pdf, color='skyblue', alpha=0.4)
plt.title(f'Chi-Square Distribution (df={df})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# F- Distribution
A common procedure in scientific experimentation is to test multiple treatments
across groupssay, different fertilizers on different blocks of a field. This is similar to
the A/B/C test referred to in the chi-square distribution , except we are dealing with measured continuous values rather than
counts. In this case we are interested in the extent to which differences among group
means are greater than we might expect under normal random variation. The F-statistic measures this and is the ratio of the variability among the group means to the
variability within each group (also called residual variability).

- The F-distribution is used with experiments and linear models involving meas
ured data.
- The F-statistic compares variation due to factors of interest to overall variation.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define parameters for the F-distribution
dfn = 5  # Degrees of freedom for the numerator
dfd = 2  # Degrees of freedom for the denominator

# Generate a range of x values for plotting
x = np.linspace(0, 5, 1000)

# Calculate the F-distribution PDF
pdf = stats.f.pdf(x, dfn, dfd)

# Plot the F-distribution
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label=f'F-Distribution (dfn={dfn}, dfd={dfd})', color='green')
plt.fill_between(x, pdf, color='lightgreen', alpha=0.5)
plt.title(f'F-Distribution (dfn={dfn}, dfd={dfd})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

# Statistical Experiments and Significance Testing

# A/B Testing
An A/B test is an experiment with two groups to establish which of two treatments,
products, procedures, or the like is superior. Often one of the two treatments is the
standard existing treatment, or no treatment. If a standard (or no) treatment is used,
it is called the control. A typical hypothesis is that a new treatment is better than the
control

## Key Terms for A/B Testing

- **Treatment**  
  Something (drug, price, web headline) to which a subject is exposed.

- **Treatment group**

# Step 2: Analyze Data 


```python
import statsmodels.api as sm

# Calculate conversion rates
conversion_rate_A = conversions_A.mean()
conversion_rate_B = conversions_B.mean()

print(f'Conversion rate for group A: {conversion_rate_A:.4f}')
print(f'Conversion rate for group B: {conversion_rate_B:.4f}')

# Perform a two-proportion z-test
count_A = conversions_A.sum()
count_B = conversions_B.sum()
nobs_A = n_A
nobs_B = n_B

z_stat, p_val = sm.stats.proportions_ztest([count_A, count_B], [nobs_A, nobs_B])
print(f'z-statistic: {z_stat:.4f}')
print(f'p-value: {p_val:.4f}')
```

# Hypothesis Tests

*P-value:* This is the probability of observing the test statistic or something more extreme if the null hypothesis is true. A small p-value (typically  ) indicates strong evidence against the null hypothesis, leading to its rejection.

*Conclusion:* Based on the p-value and the significance level, you either reject the null hypothesis (if p-value  ) or fail to reject the null hypothesis (if p-value > ).

## Hypothesis Testing

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using sample data to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

### Steps in Hypothesis Testing

#### State the Hypotheses
- **Null Hypothesis (H0)**: A statement of no effect or no difference.
- **Alternative Hypothesis (H1)**: A statement that contradicts the null hypothesis.

#### Choose the Significance Level ()
- Common choices are 0.05, 0.01, or 0.10.

#### Select the Appropriate Test Statistic
- Depends on the type of data and sample size (e.g., t-test, z-test, chi-square test).

#### Calculate the Test Statistic and P-Value
- The test statistic measures how far the sample data deviates from the null hypothesis.
- The p-value indicates the probability of obtaining a result at least as extreme as the one observed, assuming the null hypothesis is true.

#### Make a Decision
- If the p-value is less than the significance level, reject the null hypothesis.
- Otherwise, fail to reject the null hypothesis.

### Example: Two-Sample t-Test

Let's consider an example where we want to compare the mean conversion rates between two groups (A and B) to determine if there is a significant difference.

#### Step 1: State the Hypotheses
- H0: The mean conversion rate of group A is equal to the mean conversion rate of group B.
- H1: The mean conversion rate of group A is not equal to the mean conversion rate of group B.

#### Step 2: Choose the Significance Level ()
-  = 0.05

#### Step 3: Select the Appropriate Test Statistic
- We will use a two-sample t-test since we are comparing the means of two independent samples.






#### Step 4: Calculate the Test Statistic and P-Value
Here's how to perform the two-sample t-test in Python:

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Simulate data
np.random.seed(42)

# Number of users
n_A = 1000  # Number of users in group A
n_B = 1000  # Number of users in group B

# Conversion rates
p_A = 0.15  # Conversion rate for group A
p_B = 0.18  # Conversion rate for group B

# Simulate conversions (1 = converted, 0 = not converted)
conversions_A = np.random.binomial(1, p_A, n_A)
conversions_B = np.random.binomial(1, p_B, n_B)

# Perform a two-sample t-test
t_stat, p_val = stats.ttest_ind(conversions_A, conversions_B)

print(f't-statistic: {t_stat:.4f}')
print(f'p-value: {p_val:.4f}')
```

```python
# Set significance level
alpha = 0.05

# Decision based on p-value
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the groups.")
```

## Explanation of the Code

### Simulate Data
- We simulate conversion data for 1000 users in each group (A and B) with conversion rates of 0.15 and 0.18, respectively.

### Perform a Two-Sample t-Test
- `stats.ttest_ind(conversions_A, conversions_B)` performs an independent two-sample t-test.
- `t_stat` is the calculated t-statistic.
- `p_val` is the p-value.

### Decision Making
- If the p-value is less than 0.05, we reject the null hypothesis, indicating a significant difference between the conversion rates of the two groups.
- Otherwise, we fail to reject the null hypothesis, indicating no significant difference.

### Interpreting Results
- If the p-value is low (typically less than 0.05), it suggests that the observed data is unlikely under the null hypothesis, leading us to reject H0.
- If the p-value is high, it suggests that the observed data is consistent with the null hypothesis, leading us to fail to reject H0.

## The Null Hypothesis (H0)

The Null Hypothesis (H0) is a fundamental concept in statistical hypothesis testing. ### Key Points about the Null Hypothesis
- **Baseline Assumption:** The null hypothesis assumes that any observed effect in the data is due to random chance. It serves as a starting point for statistical testing.
- **Formulation:** The null hypothesis is typically formulated as a statement of equality (e.g., no difference between group means, no association between variables).
- **Testing:** Statistical tests are used to evaluate the null hypothesis. The goal is to determine whether the sample data provides sufficient evidence to reject H0.
- **Decision Rule:** The decision to reject or fail to reject the null hypothesis is based on a pre-determined significance level (), often set at 0.05. If the p-value of the test is less than , the null hypothesis is rejected.

## Permutation Test
In a permutation procedure, two or more samples are involved, typically the groups in an A/B or other hypothesis test. "Permute" means to change the order of a set of values. The first step in a permutation test of a hypothesis is to combine the results from groups A and B (and, if used, C, D, ...). This is the logical embodiment of the null hypothesis that the treatments to which the groups were exposed do not differ. We then test that hypothesis by randomly drawing groups from this combined set and seeing how much they differ from one another.

### Permutation Procedure
Combine Data:
Combine the results from the different groups into a single data set.

### Shuffle and Resample:
Shuffle the combined data and then randomly draw (without replacement) a resample of the same size as group A (clearly, it will contain some data from the other groups).

### Resample Other Groups:
From the remaining data, randomly draw (without replacement) a resample of the same size as group B. Repeat for groups C, D, and so on. You have now collected one set of resamples that mirror the sizes of the original samples.

### Calculate Statistic:
Whatever statistic or estimate was calculated for the original samples (e.g., difference in group proportions), calculate it now for the resamples and record it. This constitutes one permutation iteration.

### Repeat:
Repeat the previous steps 

R times to yield a permutation distribution of the test statistic.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
group_A = np.random.binomial(1, 0.15, 1000)
group_B = np.random.binomial(1, 0.18, 1000)

# Define test statistic function (difference in means)
def test_statistic(data1, data2):
    return np.mean(data1) - np.mean(data2)

# Observed test statistic
observed_stat = test_statistic(group_A, group_B)

# Combine data
combined = np.concatenate([group_A, group_B])

# Number of permutations
R = 10000
permuted_stats = np.zeros(R)

# Permutation procedure
for i in range(R):
    np.random.shuffle(combined)
    perm_A = combined[:len(group_A)]
    perm_B = combined[len(group_A):]
    permuted_stats[i] = test_statistic(perm_A, perm_B)

# Calculate p-value
p_value = np.sum(permuted_stats >= observed_stat) / R

# Plot permutation distribution
plt.hist(permuted_stats, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(observed_stat, color='r', linestyle='dashed', linewidth=2, label='Observed Statistic')
plt.xlabel('Test Statistic')
plt.ylabel('Frequency')
plt.title('Permutation Distribution')
plt.legend()
plt.grid(True)
plt.show()

# Print results
print(f'Observed Statistic: {observed_stat}')
print(f'P-value: {p_value}')
```

## Explanation of the Code

### Simulate Data
- We simulate conversion data for 1000 users in each group (A and B) with conversion rates of 0.15 and 0.18, respectively.

### Define Test Statistic
- The test statistic function calculates the difference in means between two groups.

### Observed Test Statistic
- We calculate the observed difference in means between group A and group B.

### Combine Data
- The data from both groups is combined into a single array.

### Permutation Procedure
1. We shuffle the combined data and split it into new groups A and B of the same sizes as the original groups.
2. We calculate the test statistic for each permutation and store it.
3. We repeat this process \( R \) times to build the permutation distribution.

### Calculate P-Value
- The p-value is calculated as the proportion of permuted test statistics that are as extreme as or more extreme than the observed test statistic.

### Plot Permutation Distribution
- We plot the permutation distribution and mark the observed test statistic on the plot.

### Interpretation
- If the p-value is low (typically less than 0.05), it suggests that the observed data is unlikely under the null hypothesis, leading us to reject \( H_0 \).
- If the p-value is high, it suggests that the observed data is consistent with the null hypothesis, leading us to fail to reject \( H_0 \).
